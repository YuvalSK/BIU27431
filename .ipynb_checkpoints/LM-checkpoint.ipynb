# Linear models
## y=bx

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size': 18})

def LM(x):
    y=[]
    for i in x:
        y.append(4.5-3.5*i)
    return y

#observations
x = [0.90,0.75,0.72,0.65,0.2]
y = [0.6,1.29,1.99,2.69,3.4] 
plt.scatter(x,y,c='r',label='y')

#predictions
y_pred = LM(x)
plt.plot(x,y_pred,label='y_p')

# Error functions
ME = np.subtract(y,y_pred).mean() 
MSE = np.square(np.subtract(y,y_pred)).mean() 
RMSE = np.sqrt(MSE)
print(f'Model Estimation:\nME: {ME:.3f}\nMSE: {MSE:.3f}\nRMSE: {RMSE:.3f}')

#bla bla for visualization
plt.grid()
plt.legend()
plt.xlabel("X [AU]")
plt.ylabel("Y [AU]")
plt.title('Linear Model')
```
