# PCA for exploring neuronal data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
plt.style.use('classic')
%matplotlib inline

df = pd.read_csv('some_data.csv',index_col=0)
neurons = df.index
plt.scatter(range(0,len(df.iloc[:,0])),df.iloc[:,0],c='r',label='t1')
plt.scatter(range(0,len(df.iloc[:,0])),df.iloc[:,1],c='b', label='t2')
plt.scatter(range(0,len(df.iloc[:,0])),df.iloc[:,2],c='g', label='t3')
plt.scatter(range(0,len(df.iloc[:,0])),df.iloc[:,3],c='k', label='t4')
plt.scatter(range(0,len(df.iloc[:,0])),df.iloc[:,4],c='gray', label='t5')

plt.xlim(0,len(df.iloc[:,1]))
plt.xlabel('Neuron [#]')
plt.ylabel('FR [sp/s]')
plt.legend()
```

## Balagan! we have the FR of 90 neurons, under 5 different conditions (tasks).
## Our goal here is to find groups of neurons, which might share similar properties 
## 1. First, let's reduce dimensions - how many dimensions (=tasks) are needed to explain most of the variance in our experiment

```python
pca = PCA()
pca.fit(df)

#how mach variance was explained by each PC
var = pca.explained_variance_ratio_
#note that the total variance should sum to 1
#print(var.sum())

fig, ax = plt.subplots(figsize=(12,10), nrows=2, ncols=1)

ax[0].scatter(range(1,6),var,label='PC')
ax[0].plot(range(1,6),var.cumsum(), marker = 'x', c='r', linestyle='--',label='Total sum')

ax[0].axhline(y=1,c='k')
ax[0].set_xlabel('Number of PCs [#]')
ax[0].set_ylabel('Variance [v]')
ax[0].set_ylim(0,1.5)
ax[0].grid()
ax[0].legend()

pca = PCA(n_components=2)
pca.fit(df)
df_pca = pca.transform(df)
ax[1].scatter(df_pca[:,0],df_pca[:,1],c='gray')
ax[1].set_title('PC space')
ax[1].grid()
for i,n in enumerate(neurons):
    ax[1].annotate(n,(df_pca[i,0],df_pca[i,1]),fontsize=8, 
        textcoords='offset points', ha='center', va='bottom', color='k')
ax[1].set_xlabel(f'PC1 [{100*var[0]:.2f}%]')
ax[1].set_ylabel(f'PC2 [{100*var[1]:.2f}%]')
```

## we can see that 2 PCs account for the vast majority of the variance (~97%!) in our experimet
### PC1 ~ 87% and PC2 ~10% 
## 2. Next, we test how many clusters are a good fit for the reduced data (in PC space)

```python
from sklearn.cluster import KMeans
fig, ax = plt.subplots(figsize=(12,10), nrows=2, ncols=1)
wcss = []
centroids=[]
ks=10
for i in range(1,ks):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df_pca)
    wcss.append(kmeans.inertia_)
    if i==4:
        centroids  = kmeans.cluster_centers_
        centroids_x = centroids[:,0]
        centroids_y = centroids[:,1]
        ax[1].scatter(centroids_x,centroids_y,marker = "o", s=150,linewidths = 5, zorder = 10,c=['b','r','navy','y'])
        ax[1].scatter(df_pca[:,0],df_pca[:,1], facecolors='none', edgecolors='gray')
        ax[1].set_xlabel('PC1')
        ax[1].set_ylabel('PC2')
        ax[1].set_title(f'PC space with {i} clusters')      

ax[0].plot(range(1,ks), wcss, marker='o', linestyle='--')
ax[0].set_xlabel('# of clusters')
ax[0].set_ylabel('wcss')
ax[0].set_title('Elbow method')
ax[0].set_xlim(0,ks)
ax[0].axvline(x=4,c='k')
```
